```python
import pandas as pd
```

```python
from LeGenderary.leGenderary import leGenderary
```

```python

options = { 'male'          : 0, 
            'female'        : 1,
            'androgynous'   : 'androgynous',
            'unknown'       : 'unknown',
            'maleConfirm'   : 'male needs confirmation',
            'femaleConfirm' : 'female needs confirmation',
            'dict1'         : '../dicionarios/dict1.txt',
            'dict2'         : '../dicionarios/custom.txt',
            'customDict'    : '../dicionarios/custom.txt',
            'bingAPIKey'    : 'ABC123478ZML'
          }

gender      = leGenderary(options)
```

```python
cursos = ['eng_comp','jogos','proc_dados', 'lic_inf', 'lic_comp','si']
```

```python
def get_gender(row):
    first_name = row['Nome'].split()[0]
    return gender.determineFromDictionary(first_name)
```

```python
gender.determineFromDictionary('lahys')
```

```python
df_cursos = pd.DataFrame()
```

```python
for curso in cursos:
    df = pd.read_csv('../outputs/dfs_cursos/'+curso+'.csv', index_col=0)
    df['genero'] = df.apply(get_gender, axis=1)
    df['curso'] = curso
    df_cursos = df_cursos.append(df, ignore_index=True)
```
